```python
#Loading Library
import seaborn as sns
import matplotlib.pyplot as plt

#Loading Dataset
ship = sns.load_dataset('titanic')

sns.set(style='whitegrid')
sns.boxplot(x='class', y='fare',data=ship)
plt.title('Testing Success')
ship.describe()
```

```python
# Single axis data
sns.boxplot(x=ship['survived']);
```

```python
# Box plot example with differnt paramters
sns.boxplot(x='survived',y='age',hue='parch',data=ship,palette='Set1',dodge=True)
plt.title('Sunk Ship Analysis')
plt.show()
```

```python
# changing color
sns.boxplot(x='survived',y='age',hue='parch',data=ship,color='#3ceb10')
plt.title('Sunk Ship Analysis')
plt.show()
```

```python
# Assignment (We can change the colors with the help of palette option)
sns.boxplot(x='survived',y='age',hue='parch',data=ship,palette='Set2',dodge=True)
plt.title('Sunk Ship Analysis')
plt.show()
```
